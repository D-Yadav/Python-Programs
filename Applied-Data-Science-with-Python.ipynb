```python
One sample t test -
```

```python
Null hypothesis -there is no significant difference between the mean mean of population 1 = mean of pupulation 2
Alternative hypothesis -there is a significant difference mean pf p1 != mean of p2
```

```python
popmean = 120

sample2 = [345, 195, 245, 486, 390, 295, 298, 1234, 199]
```

```python
from scipy import stats

t_stats, p_value =stats.ttest_1samp(sample2, popmean=popmean)
```

```python
print(p_value)
print(t_stats)
#if the p value coming as less than of equal to 0.05 than we will reject null hypothesis 
#otherwise we will accept null hypothesis.

#higher the t stats lower the p value more evidance to reject null hypothesis
# if the t stats value is near to zero than your p value will be higher and there will be very less evidance to reject null hypothesis
```

```python
t_stats = m-s/n

m =population mean
s = std of group
n = size of group
```
